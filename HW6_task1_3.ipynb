```python
# 1) Известно, что генеральная совокупность распределена нормально
# со средним квадратическим отклонением, равным 16.
# Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,
# если выборочная средняя M = 80, а объем выборки n = 256.
import numpy as np
import scipy.stats as stats
M=80
n=256
z=stats.norm.ppf(0.975)
x_1 = M-z*(16/np.sqrt(256))
print(f'Нижняя граница интервала равна {x_1}')
x_2 =  M+z*(16/np.sqrt(256))
print(f'Верхняя граница интервала равна {x_2}')
print('С вероятностью 95% интервал (78,04;81,96) будет содержать математическое ожидание нормально распределенной совокупности')
```
